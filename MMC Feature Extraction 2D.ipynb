```python
import warnings
warnings.filterwarnings("ignore")
```

```python
'''------Reading files------'''
import os

'''------Audio preprocessing------'''
from pydub import AudioSegment

'''------Manipulating data------'''
import numpy as np
from sklearn.preprocessing import LabelEncoder # Numerical encoding

'''------Loading the audio------'''
import librosa
import librosa.display
from IPython import display
```

## 2. Load Processed Data

```python
dir25_1 = "Dataset25/MER_25"
data25_1 = os.listdir(dir25_1)
file_dir25_1 = []
count1 = 0

for mooddir25 in data25_1:
    if(not r'.' in mooddir25):
        moodDirName25 = os.path.join(dir25_1,mooddir25)
        mooddir25=np.array(os.listdir(moodDirName25))
        for audio in mooddir25:
            if(audio[-3:]=='mp3'):
                count1+=1
                fn=os.path.join(moodDirName25,audio)
                file_dir25_1.append(fn)
                print(str(count1),fn)
```

```python
dir25_2 = "Dataset25/MER_Extra25"
data25_2 = os.listdir(dir25_2)
file_dir25_2 = []
count2 = 0

for mooddir25 in data25_2:
    if(not r'.' in mooddir25):
        moodDirName25 = os.path.join(dir25_2,mooddir25)
        mooddir25=np.array(os.listdir(moodDirName25))
        for audio in mooddir25:
            if(audio[-3:]=='mp3'):
                count2+=1
                fn=os.path.join(moodDirName25,audio)
                file_dir25_2.append(fn)
                print(str(count2),fn)
```

## 3. Feature extraction

```python
def feature2d(file):
    try:
        y,sr = librosa.load(file)
    except:
        print('No such file')
        quit()
            
    # Spectrogram
    stft = librosa.stft(y)
    stft_db = librosa.amplitude_to_db(abs(stft))
    spec = stft_db
    # MFCC
    mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=20)
    # Mel
    s = librosa.feature.melspectrogram(y=y, sr=sr)
    mel = librosa.amplitude_to_db(s, ref=np.max)

    return spec, mfccs, mel
```

```python
n1 = len(file_dir25_1) # num of audio data
spec1 = np.empty([n1, 1025, 1099])
mfccs1 = np.empty([n1, 20, 1099])
mel1 = np.empty([n1, 128, 1099])
```

```python
# Label the mood
mood1 = []
for i in range(len(file_dir25_1)):
    m = os.path.basename(os.path.dirname(file_dir25_1[i]))
    mood1.append(m)
```

```python
mood1t = mood1
for i in range(len(mood1)):
    if mood1[i] == "Q1":
        mood1t[i] = "happy"
    
    elif mood1[i] == "Q2":
        mood1t[i] = "angry"
    
    elif mood1[i] == "Q3":
        mood1t[i] = "sad"
    
    elif mood1[i] == "Q4":
        mood1t[i] = "relaxed" 
```

```python
for i in range(len(file_dir25_1)):
    spec1[i], mfccs1[i], mel1[i] = feature2d(file_dir25_1[i])
```

```python
# save memory usage
spec1 = spec1.astype(np.float32)
mfccs1 = mfccs1.astype(np.float32)
mel1 = mel1.astype(np.float32)
```

```python
n2 = len(file_dir25_2) # num of audio data
spec2 = np.empty([n2, 1025, 1099])
mfccs2 = np.empty([n2, 20, 1099])
mel2 = np.empty([n2, 128, 1099])
```

```python
# Label the mood
mood2 = []
for i in range(len(file_dir25_2)):
    m = os.path.basename(os.path.dirname(file_dir25_2[i]))
    mood2.append(m)
```

```python
for i in range(len(file_dir25_2)):
    spec2[i], mfccs2[i], mel2[i] = feature2d(file_dir25_2[i])
```

```python
# save memory usage
spec2 = spec2.astype(np.float32)
mfccs2 = mfccs2.astype(np.float32)
mel2 = mel2.astype(np.float32)
```

## 4. Label Encoding

```python
np.unique(mood1t)
```

```python
le = LabelEncoder()
moodtrans1=le.fit(np.array(mood1t))
moodtrans1=le.transform(np.array(mood1t))
np.unique(moodtrans1)
```

```python
np.unique(mood2)
```

```python
moodtrans2=le.fit(np.array(mood2))
moodtrans2=le.transform(np.array(mood2))
np.unique(moodtrans2)
```

## 5. Save 2D features into npz. file

```python
#Save the 2D features and labels
np.savez_compressed("Features2D/npz/Features1.npz", 
                    spec = spec1, 
                    mfcc = mfccs1,
                    mel = mel1,
                    target = moodtrans1)

np.savez_compressed("Features2D/npz/Features2.npz", 
                    spec = spec2, 
                    mfcc = mfccs2,
                    mel = mel2,
                    target = moodtrans2)
```
